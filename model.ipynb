```python
x=df.iloc[:,0].values
```

```python
x[0]
```

```python

import string
all_words=[]
punctuation=string.punctuation+'\n'

for i in x:
    i=i.lower()
    i=i.strip().split(" ")
    for j in i:
        if j not in stopwords and j not in punctuation:
            all_words.append("".join([k for k in j if k not in punctuation]))
len(all_words)
```

```python
word_frequencies={}

for i in all_words:
    if i not in word_frequencies:
        word_frequencies[i]=1
    else:
        word_frequencies[i]+=1
```

```python
word_frequencies
```

```python
max_frequency=max(word_frequencies.values())
```

```python
normalized_frequencies={}
for i in word_frequencies:
    normalized_frequencies[i]=word_frequencies[i]/max_frequency
```
